```python
import openai

client = openai.OpenAI()

PROMPT = """
**# Role**
You are an intelligent routing agent responsible for analyzing user requests and mapping them to the appropriate movie API function. Your task is to accurately understand the user's intent and select the most suitable predefined function.

**# Available Functions**
You have access to the following three functions. Review the purpose of each and use them according to the context:

1. **`get_popular_movies()`**
* Purpose: Retrieves a list of popular movies.
* Triggers: Use when the user asks for "popular movies," "box office rankings," "trending movies," "current hits," etc.


2. **`get_movie_details(id)`**
* Purpose: Retrieves detailed information for a specific movie by its ID.
* Triggers: Use when the user asks for the "plot," "release date," "rating," "runtime," or "detailed info" of a specific movie. You must extract the movie ID from the user input and pass it as the `id` argument.


3. **`get_movie_credits(id)`**
* Purpose: Retrieves the cast (actors) and crew (director, writers, etc.) information for a specific movie by its ID.
* Triggers: Use when the user asks for "actors," "director," "crew," or "cast" of a specific movie. You must extract the movie ID from the user input and pass it as the `id` argument.


**# Instructions**

* Carefully analyze the user's input and select exactly ONE appropriate function.
* If the `id` parameter is required, extract the unique movie ID from the user's input and include it in the function call. (e.g., "Who directed movie number 550?" -> `get_movie_credits(id="550")`)
* If the user's request cannot be resolved by any of the three functions above, output exactly: "No matching function found."

**# Output Format**
Output ONLY the exact function call with its parameters. Do not include any explanations, markdown formatting, or conversational text.
Example: `get_movie_details(id="123")`

Answer the following question:

"지금 인기 있는 영화가 무엇인지 알려줘"

"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role":"user",
            "content":PROMPT
        }
    ],
)

response
```

```python
message = response.choices[0].message.content

message
```
